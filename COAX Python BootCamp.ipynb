### 1. There is string s = "Python Bootcamp". Write the code that hashes string.

```python
import hashlib
```

```python
algorithms_list = list(hashlib.algorithms_available)
print(algorithms_list)
```

```python
def hash_func(text, algorithm):
    hash_object = eval( 'hashlib.' + algorithm + "({})".format(text))
    print(algorithm + '\t' + hash_object.hexdigest() + '\n')
    
```

```python
for algorithm in algorithms_list:
    try:
        hash_func(b'Python Bootcamp', algorithm)
    except:
        print(algorithm + ' is unavailable.' + '\n')
```
